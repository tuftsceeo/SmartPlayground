````python
# PyFluidSynth Example

import time
import numpy
import pyaudio
import fluidsynth

pa = pyaudio.PyAudio()
strm = pa.open(
    format = pyaudio.paInt16,
    channels = 2, 
    rate = 44100, 
    output = True)

s = []

fl = fluidsynth.Synth()

# Initial silence is 1 second
s = numpy.append(s, fl.get_samples(44100 * 1))

sfid = fl.sfload("FluidR3_GM.sf2")
fl.program_select(0, sfid, 0, 0)

fl.noteon(0, 60, 30)
fl.noteon(0, 67, 30)
fl.noteon(0, 76, 30)

# Chord is held for 2 seconds
s = numpy.append(s, fl.get_samples(44100 * 2))

fl.noteoff(0, 60)
fl.noteoff(0, 67)
fl.noteoff(0, 76)

# Decay of chord is held for 1 second
s = numpy.append(s, fl.get_samples(44100 * 1))

fl.delete()

samps = fluidsynth.raw_audio_string(s)

print(len(samps))
print('Starting playback')
strm.write(samps)
print("Done")
strm.close()
````

````python
#Fluidsynth Example for Sequencer using custom callbacks

#!/usr/bin/env python3

import time
import fluidsynth
from ctypes import *

seqduration = 1000

def schedule_next_callback():
    # I want to be called back before the end of the next sequence
    callbackdate = int(now + seqduration/2)
    print("Call back date: ", now)
    sequencer.timer(callbackdate, dest=mySeqID)

def schedule_next_sequence():
    global now
    # the sequence to play
    print("Load Next: ", now)
    # the beat : 2 beats per sequence
    sequencer.note(int(now + seqduration * 1/4), 0, 60, duration=250, velocity=80, dest=synthSeqID)
    sequencer.note(int(now + seqduration * 2/4), 0, 60, duration=250, velocity=80, dest=synthSeqID)
    sequencer.note(int(now + seqduration * 3/4), 0, 60, duration=250, velocity=80, dest=synthSeqID)
    sequencer.note(int(now + seqduration * 4/4), 0, 60, duration=250, velocity=80, dest=synthSeqID)
    
    # melody
    # sequencer.note(int(now + seqduration*1/10), 10, 45, duration=250, velocity=int(127*2/3), dest=synthSeqID)
    # sequencer.note(int(now + seqduration*4/10), 10, 50, duration=250, velocity=int(127*2/3), dest=synthSeqID)
    # sequencer.note(int(now + seqduration*8/10), 10, 55, duration=250, velocity=int(127*3/3), dest=synthSeqID)
    # so that we are called back early enough to schedule the next sequence
    schedule_next_callback()
    
    now = now + seqduration
    print("End Time:", now)

def seq_callback(time, event, seq, data):
    schedule_next_sequence()

if __name__=="__main__":
    global sequencer, fs, mySeqID, synthSeqID, now
    fs = fluidsynth.Synth()
    fs.start(driver='alsa')
    # you might have to use other drivers:
    #fs.start(driver="alsa", midi_driver="alsa_seq")

    sfid = fs.sfload("FluidR3_GM.sf2")
    fs.program_select(0, sfid, 128, 0)
    fs.program_select(10, sfid, 0, 0) # use the same program for channel 2 for cheapness

    sequencer = fluidsynth.Sequencer()
    synthSeqID = sequencer.register_fluidsynth(fs)
    mySeqID = sequencer.register_client("mycallback", seq_callback)
    now = sequencer.get_tick()
    schedule_next_sequence()
    
    time.sleep(10)
    
    sequencer.delete()
    fs.delete()
    print("done")
````

# Midi Test Run

````python
# Actual playback of MIDI File Drums with Adjustable Tempo

#!/usr/bin/env python3

import time
import fluidsynth
import mido
import random
from ctypes import *

""" 
============================
Adding extra wrappers for fluidsynth c library 
============================
"""
fluid_event_control_change = fluidsynth.cfunc('fluid_event_control_change', None,
                         ('evt', c_void_p, 1),
                         ('channel', c_int, 1),
                         ('control', c_short, 1),
                         ('val', c_int, 1))

def control_change(seq, time, channel, control, val, source=-1, dest=-1, absolute=True):
        evt = seq._create_event(source, dest)
        fluid_event_control_change(evt, channel, control, val)
        seq._schedule_event(evt, time, absolute)
        fluidsynth.delete_fluid_event(evt)
        

""" 
============================
Setup Midi 
============================
"""
#seqduration = 1000

midi_file = "audio/drums120.mid"
mid = mido.MidiFile(midi_file)
track = mid.tracks[0] # Only take the first track
print("File Duration (seconds): ", mid.length)

""" 
============================
Playback Functions
============================
"""

def schedule_next_callback():
    # I want to be called back before the end of the next sequence
    callbackdate = int(now + seqduration/2)
    print("Call back date: ", now)
    sequencer.timer(callbackdate, dest=mySeqID)   
    
def load_midi_to_sequencer():

    global now
    print("Load Next: ", now)
    schedule_next_callback() #start the next callback now before the delay of reading midi
    
    for msg in track:
        # print(msg)
        now += msg.time

        if msg.type == 'note_on':
            sequencer.note_on(now, channel=0, key=msg.note, velocity=msg.velocity, dest=synthSeqID)
            #print("Time:", now, 
            #      "  Channel:", msg.channel, 
            #      "  Key:", msg.note, 
            #      "  Velocity:", msg.velocity)
            
        elif msg.type == 'control_change':
            control_change(sequencer, now, channel=0, control=msg.control, val=msg.value, source=-1, dest=synthSeqID)

                
    
    print("End Time:", now)

def seq_callback(time, event, seq, data):
    #schedule_next_sequence()
    load_midi_to_sequencer()
    
""" 
============================
Main
============================
"""

if __name__=="__main__":

    global sequencer, fs, mySeqID, synthSeqID, now, seqduration
    fs = fluidsynth.Synth()
    fs.start(driver='alsa')
    # you might have to use other drivers: alsa is for RPi
    

    sfid = fs.sfload("FluidR3_GM.sf2")
    fs.program_select(0, sfid, 128, 0)
    fs.program_select(10, sfid, 0, 0) # use the same program for channel 2 for cheapness

    sequencer = fluidsynth.Sequencer()
    time_scale = 200 #ticks per second
    seqduration = mid.length * 1000
    
    synthSeqID = sequencer.register_fluidsynth(fs)
    fluidsynth.fluid_sequencer_set_time_scale(sequencer.sequencer, time_scale)
    mySeqID = sequencer.register_client("mycallback", seq_callback)
    
    now = sequencer.get_tick()
      
    load_midi_to_sequencer()
    
    start_time = time.time()
    
    while time.time()-start_time<20:
        #scale = random.randint(1,5)
        #time_scale = scale * 500 #ticks per second
        #seqduration = mid.length * time_scale
        #fluidsynth.fluid_sequencer_set_time_scale(sequencer.sequencer, time_scale)
        #print("New Time Scale:", time_scale)
        time.sleep(3)
    
    sequencer.delete()
    fs.delete()
    print("done")
````
